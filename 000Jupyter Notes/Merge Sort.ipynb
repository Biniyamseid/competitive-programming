```python
class Solution:
    def sortArray(self, nums):
        return self.merge_sort(nums)
    
    
    def merge_sort(self,arr):
        size = len(arr)
        sorted_arr = [0] * size
        width = 1

        while width < size:
            for i in range(0, size, 2 * width):
                self.merge(arr, sorted_arr, i, min(i + width, size), min(i + 2 * width, size))
            arr, sorted_arr = sorted_arr, arr
            width *= 2

        return arr


    def merge(self,arr, sorted_arr, start, mid, end):
        left = start
        right = mid
        index = start

        while left < mid and right < end:
            if arr[left] <= arr[right]:
                sorted_arr[index] = arr[left]
                left += 1
            else:
                sorted_arr[index] = arr[right]
                right += 1
            index += 1

        while left < mid:
            sorted_arr[index] = arr[left]
            left += 1
            index += 1

        while right < end:
            sorted_arr[index] = arr[right]
            right += 1
            index += 1
            
            
            
            
            
                
    def merge_sort2(self,nums):
        # bottom cases: empty or list of a single element.
        if len(nums) <= 1:
            return nums

        pivot = int(len(nums) / 2)
        left_list = self.merge_sort2(nums[0:pivot])
        right_list = self.merge_sort2(nums[pivot:])
        return self.merge2(left_list, right_list)


    def merge2(self,left_list, right_list):
        left_cursor = right_cursor = 0
        ret = []
        while left_cursor < len(left_list) and right_cursor < len(right_list):
            if left_list[left_cursor] < right_list[right_cursor]:
                ret.append(left_list[left_cursor])
                left_cursor += 1
            else:
                ret.append(right_list[right_cursor])
                right_cursor += 1

        # append what is remained in either of the lists
        ret.extend(left_list[left_cursor:])
        ret.extend(right_list[right_cursor:])

        return ret
```
